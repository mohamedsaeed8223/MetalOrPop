```python
import pandas as pd
```

```python
dataset = pd.read_excel("Egyptian Tweets.xlsx")
```

```python
# prompt: crop the first 2 columns for dataset and save it dataset

dataset = dataset.iloc[:, :2]
dataset.head()
```

```python
# prompt: Using dataframe dataset: change the name of the second column to "translation" and make all of its values null

dataset.rename(columns={dataset.columns[1]: 'Translation'}, inplace=True)
dataset['Translation'] = None
dataset.head()
```

```python
# prompt: now crop dataset to the last 500 rows and remove the original indexing and save the dataset as an excel file

dataset = dataset.tail(500).reset_index(drop=True)
dataset.to_excel("cropped_dataset.xlsx")
```

```python
# prompt: split the dataset to 5 equal parts and save each to its own file and remove the indexing in them

import pandas as pd

# Split the dataset into 5 equal parts
df_list = [dataset.iloc[i:i + len(dataset) // 5] for i in range(0, len(dataset), len(dataset) // 5)]

# Save each part to its own file and remove the indexing
for i, df in enumerate(df_list):
  df.reset_index(drop=True, inplace=True)
  df.to_excel(f"part{i+1}.xlsx")
```
